# Thao tác chuỗi đơn giản

```python
# Nhập 1 chuỗi bất kì
String  = input("Nhập 1 chuỗi bất kì: ")

# In hoa chuỗi
UpperString = String.upper()

# In thường chuỗi
LowerString = String.lower()

# In độ dài chuỗi
LengthString = len(String)

# Kết quả
print(f"Chuỗi in hoa: {UpperString}")
print(f"Chuỗi in thường: {LowerString}")
print(f"Độ dài chuỗi: {LengthString}")
```
