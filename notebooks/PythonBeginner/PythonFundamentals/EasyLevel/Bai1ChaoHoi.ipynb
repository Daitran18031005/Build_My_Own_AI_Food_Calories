## Chào hỏi và năm sinh

```python
from datetime import datetime #module datetime để lấy năm hiện tại
```

```python
Name = input("Nhập tên của bạn: ")
BornYear = int(input("Nhập năm sinh của bạn: "))
CurrentYear = datetime.now().year #lấy năm hiện tại từ hệ thống
Age = CurrentYear - BornYear
print(f"Chào bạn {Name}, bạn {Age} tuổi.")
```
