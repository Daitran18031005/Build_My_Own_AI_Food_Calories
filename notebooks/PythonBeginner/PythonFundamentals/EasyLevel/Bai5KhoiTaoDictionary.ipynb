# Khởi tạo Dictionary

```python
# Khởi tạo 1 dictionary person với các key: name, age, city
person = {
    "name" : "Trần Quang Đại", 
    "age" : 20,
    "city" : "Thành phố Hồ Chí Minh"
}

# In ra giá trị của từ khóa city trong dictionary person
print(person["city"]) # Cách 1
print(person.get("city")) # Cách 2
print(person.setdefault("city")) # Cách 3
# In ra toàn bộ dictionary person
print(person)
# In ra toàn bộ các key trong dictionary person
print(person.keys())
# In ra toàn bộ các value trong dictionary person
print(person.values())
# In ra toàn bộ các cặp key-value trong dictionary person
print(person.items())

# Thêm 1 từ khóa mới là job với giá trị bất kì
person["job"] = "Student"
print(person)
```
