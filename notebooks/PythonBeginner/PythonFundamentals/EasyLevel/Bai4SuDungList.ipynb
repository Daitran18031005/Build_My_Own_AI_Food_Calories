# Sử dụng list cơ bản

```python
# Khởi tạo 1 list cơ bản
ListFruits = ['Apple', 'Banana', 'Cherry', 'Orange']

# In ra phần tử đầu tiên và phần tử cuối cùng của list
print(f"Danh sách phần tử trong ListFruits: {ListFruits}")
print(f"Phần tử đầu tiên: {ListFruits[0]}")
print(f"Phần tử cuối cùng: {ListFruits[-1]}")

# Thay thế phần tử thứ hai bằng 'Grape'
ListFruits[1] = 'Grape'
print(f"Danh sách phần tử sau khi thay thế: {ListFruits}")
```
