# Cập nhật Dictionary

```python
# Cho Dictionary
Inventory = {
    "Apple": 5,
    "Banana": 10
}

# Cập nhật số lượng Banana lên 15
Inventory["Banana"] = 15
print(Inventory)

# Thêm mục mới Orange với số lượng 7
Inventory["Orange"] = 7

print(Inventory)

# In toàn bộ keys trong dictionary
print(f"Keys in Inventory: {list(Inventory.keys())}")
# In toàn bộ values tỏng dictionary
print(f"Values in Inventory: {list(Inventory.values())}")
# In toàn bộ cặp key-value trong dictionary
print(f"Items in Inventory: {list(Inventory.items())}")
```
