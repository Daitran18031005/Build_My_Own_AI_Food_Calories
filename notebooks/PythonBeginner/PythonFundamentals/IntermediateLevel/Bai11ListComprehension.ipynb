# List Comprehension với điều kiện

```python
#  Sử dụng list comprehension để tạo một list numbers = [1, 6, 11, 16, 21, 26]
Numbers = [i for i in range(1, 30, 5)]
print(f"Danh sách nummbers: {Numbers}")

# Numbers chỉ chứa các số lớn hơn 10 và chia hết cho 5
FilteredNumbers = [i for i in Numbers if i > 10 and i % 5 == 0]
print(f"Danh sách các số trong Numbers lớn hơn 10 và chia hết cho 5: {FilteredNumbers}")
```

# List comprehension
Cú pháp:

NewList = [expression for item in iterable]

expression: giá trị được thêm vào list, số, phép biến đổi, hàm,...

for item in iterable: vòng lặp duyệt iterable
