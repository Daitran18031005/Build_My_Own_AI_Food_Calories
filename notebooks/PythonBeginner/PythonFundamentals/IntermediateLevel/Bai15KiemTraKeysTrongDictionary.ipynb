# Kiểm tra Keys trong Dictionary

```python
Config = {
    "Port": 8080,
    "Host": "localhost",
    "Timeout": 30
}

Config["User"] = "Đại"

# Kiểm tra xem keys user có tồn tại trong Config không và in ra thông báo
if "User" in Config:
    print(f"Key 'User' có tồn tại trong Config với giá trị: {Config["User"]}")
else:
    print(f"Key 'User' không tồn tại trong Config")
```
