# Thống kê chữ cái

```python
# Nhận 1 chuỗi
String = input("Nhập một chuỗi: ")

# Độ dài chuỗi
StringLength = len(String)

LowerCount = 0
UpperCount = 0
NumberCount = 0

# Đếm số ký tự chữ thường và in ra
for LowerChar in String:
    if LowerChar.islower():
        LowerCount += 1

# Đếm số ký tự chữ hoa và in ra
for UpperChar in String:
    if UpperChar.isupper():
        UpperCount += 1

# Đếm số ký tự số và in ra
for NumberChar in String:
    if NumberChar.isdigit():
        NumberCount += 1

# Đếm số ký tự đặc biệt và in ra
SpecialCount = StringLength - (LowerCount + UpperCount + NumberCount)

# Kết quả
print(f"Chuỗi ký tự là: {String}")
print(f"Số ký tự chữ thường: {LowerCount}")
print(f"Số ký tự chữ hoa: {UpperCount}")
print(f"Số ký tự số: {NumberCount}")
print(f"Số ký tự đặc biệt: {SpecialCount}")
```
