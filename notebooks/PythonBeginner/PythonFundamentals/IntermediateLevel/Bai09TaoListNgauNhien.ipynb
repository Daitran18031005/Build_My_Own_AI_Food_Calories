# Tạo list ngẫu nhiên

```python
# Import module random
import random
# tạo một list chứa 10 số nguyên ngẫu nhiên từ 1 đến 50
RandomNumber = [] # Khởi tạo list rỗng
for _ in range(10): # Lặp 10 lần cho 10 số trong list
    RandomNumber.append(random.randint(1, 50)) # Thêm số ngẫu nhiên vào list

print(f"Danh sách các số ngẫu nhiên trong list: {RandomNumber}")
```
