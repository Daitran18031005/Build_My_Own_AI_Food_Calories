# Thao tác với Set

```python
Set1 = {'A', 'B', 'C', 'D'}
Set2 = {'C', 'D', 'E', 'F'}

# In ra các phần tử chỉ có trong Set1 hoặc chỉ có trong Set2
SymmetricDifference = Set1.symmetric_difference(Set2)

# Phần tử chỉ có trong Set1
OnlyInSet1 = Set1 - Set2

# Phần tử chỉ có trong Set2
OnlyInSet2 = Set2 - Set1

# In ra kết quả
print(f"Các phần tử chỉ có trong Set1 hoặc chỉ có trong Set2: {SymmetricDifference}")
print(f"Các phần tử chỉ có trong Set 1: {OnlyInSet1}")
print(f"Các phần tử chỉ có trong Set 2: {OnlyInSet2}")
```
