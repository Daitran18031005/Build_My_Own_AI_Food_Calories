# Tìm phần tử trùng lặp 

```python
Numbers = [1, 2, 3, 4, 5, 3, 2, 4, 7, 8, 10, 23, 23, 34, 11]
Numbers.sort()

# Cách 1: Sử dụng set comprehension để tách duplicates
Duplicates = {Number for Number in Numbers if Numbers.count(Number) > 1}
print(Numbers)
print(Duplicates)

# Cách 2 sử dụng 2 list để tách duplicate elements và unique elements
NewList = []
DupList = []
for Number in Numbers:
    if Number not in NewList:
        NewList.append(Number)
    else:
        DupList.append(Number)
print(Numbers)
print(NewList)
print(DupList)

```
