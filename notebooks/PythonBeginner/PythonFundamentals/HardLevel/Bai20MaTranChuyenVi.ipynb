# Ma trận chuyển vị

```python
Matrix = [[1, 2, 3], [4, 5, 6]] # Ma trận 2X3

# Chuyển vị ma trận
def Tranpose(Matrix):
    Rows = len(Matrix) # Lấy số dòng của matrix
    Cols = len(Matrix[0]) # Lấy sô cột theo dòng đầu của matrix
    Result = [[0 for _ in range(Rows)] for _ in range (Cols)] # Tạo ma trận toàn 0 để làm bộ đệm

    for i in range (Rows): # Duyệt qua các phần tử trên dòng
        for j in range (Cols): # Duyệt qua các phần tử trên cột
            Result[j][i] = Matrix[i][j] # Đảo vị trí của phần tử trong matrix
    
    return Result

TranposedMatrix = Tranpose(Matrix)
print(TranposedMatrix)
```
