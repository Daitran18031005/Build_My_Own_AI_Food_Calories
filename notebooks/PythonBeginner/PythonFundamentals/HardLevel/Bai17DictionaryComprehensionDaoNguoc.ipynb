# Dictionary Comprehension đẩo ngược

```python
Capitals = {
    "Vietnam" : "Hanoi",
    "Thailand" : "Bangkok",
    "France" : "Paris"
}

# Tạo một dictionary mới mà trong đó sử dụng dictionary comprehension để key và value đổi chỗ với nhau
NewCapitals = {City: Country for Country, City in Capitals.items()}
print(NewCapitals)

# Cách thông thường
NewCapitals2 = {}
for country, city in Capitals.items():
    NewCapitals2[city] = country
print(NewCapitals2)
```
